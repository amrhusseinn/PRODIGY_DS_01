```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Task1.csv')
df.head()
```

```python
New_df= df.drop(['Country Code','Indicator Name','Indicator Code'], axis=1)
```

```python
chosen_country = 'United States' # Chosen country
country_data = New_df[New_df['Country Name'] == chosen_country] # Filter data for the chosen country
years = country_data.columns[2:].astype(int) # Extract years and population data
population = country_data.iloc[:, 2:].values.flatten()
plt.figure(figsize=(5, 3))
plt.bar(years, population, color='black')  # Create a bar plot with years and population
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')  # Title of the plot
plt.show()
```

```python
plt.hist(population, bins=30, color='gold', edgecolor='black')
plt.show()
```
